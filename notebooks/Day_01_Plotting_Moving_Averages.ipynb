```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#select ticker
ticker="RELIANCE.NS"
data = yf.download(ticker, start ="2025-01-01", end = "2026-03-09")
```

```python
#20 day moving average
data['MA20'] = data['Close'].rolling(window =20).mean()
```

```python
plt.figure(figsize=(10,5))
plt. plot(data['Close'], label ='Close Price')
plt.plot(data['MA20'], label = '20-Day MA')
plt.title(f"{ticker} Analysis - Day1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create 2 subplots (2 rows, 1 column)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

# Plot Price and Moving Average on Top
ax1.plot(data['Close'], label='Close Price')
ax1.plot(data['MA20'], label='20-Day MA')
ax1.set_title(f"{ticker} Price & Volume Analysis")
ax1.legend()

# Plot Volume on Bottom
ax2.bar(data.index, data['Volume'].values.flatten(), color='gray', alpha=0.3, label='Volume')
ax2.set_ylabel('Volume')

plt.show()
```

#The LAG


  The orange line follows the blue line, but with a delay. This is called **Lag**.

  When the Blue line is above the Orange line, the stock is in a short-term uptrend.

  When the Blue line is below the Orange line, it’s in a downtrend.

  Quant insight: Moving averages are "Trend Following" indicators. They don't predict the future; they confirm the present.



# Mean Reversion

The blue line constantly "hugs" or returns to the orange line.

In Finance, prices often return to their average over time. This is called Mean Reversion.

Quant insight: When the gap between the blue and orange lines becomes "too wide" (mathematically defined by Standard Deviation), a Quant might bet that the price will soon "snap back" to the average.


# The "Crossover" Signal
On May 2025 the Blue line aggressively crosses above the Orange line, and then a massive rally follows.

Quant insight: This "Golden Cross" is one of the oldest algorithmic signals. If you had written a script to "Buy when Blue > Orange," you would have captured that entire climb from roughly 1250 to 1500.

In quantitative trading, a Golden Cross strategy is a "Trend Following" system. It operates on a simple logic: you buy when momentum turns positive and sell when that momentum dies.

Here is exactly when you would buy and sell using the script logic from Day 1:

1. When to Buy? (The Entry)
You enter a Long Position (Buy) when the short-term moving average (Blue) crosses above the long-term moving average (Orange).

The Confirmation: Professional quants rarely buy at the exact moment of the cross. They wait for the candle to close above the crossover point to ensure it isn't a "fake out" (whipsaw).

The "Plus" Factor: To increase your success rate, check if Volume is also increasing. If the price crosses up but volume is low, the move might lack the "fuel" to keep going.

2. Till When to Hold? (The Ride)
You stay in the trade as long as the Blue line stays above the Orange line.

During this phase, the moving averages act as dynamic support. Even if the price dips slightly, as long as the short-term average remains higher than the long-term one, the "Quant signal" remains bullish.

In your Reliance graph, this would have been the entire period from May 2025 to roughly January 2026.

3. When to Sell? (The Exit)
There are three main ways a quant decides to exit this specific trade:

The Death Cross (Standard): This is the exact opposite of your buy signal. You sell when the short-term MA crosses back below the long-term MA. This confirms the uptrend has officially ended.

Trailing Stop-Loss: Instead of waiting for a full "Death Cross" (which can be slow), you sell if the price drops a certain percentage (e.g., 5%) below the current moving average. This protects your profits.

Profit Targets: You can set a goal based on a previous "Resistance Level" (a historical high price where the stock struggled before).

```python
Action,Logic Condition,Market Meaning
BUY,Blue Line > Orange Line,Short-term buyers are now stronger than long-term sellers.
HOLD,Blue Line remains > Orange,"The market has ""upward momentum""."
SELL,Blue Line < Orange Line,"The ""Death Cross""—the trend has reversed to bearish."
```
